```python
def stringcol(df):
    #finding string cols
    strcol=[]
    for i in df.columns:
        try:
            df[i].astype(float)
        except:
            strcol.append(i)
    return strcol

def cou(df,strcol):
    #finding the count of str cols
    d=dict()
    c=0
    for i in strcol:
        c=len(set(list(df[i])))
        d[i]=c
    return d

def leb(df,strcol,k):
    #droping the str cols with no. of different values over k
    strleb=[]
    for i in strcol:
        c=len(set(list(df[i])))
        if c<k and c>1:
            strleb.append(i)
    return strleb

def labelen(df,strleb):
    #labeling
    from sklearn.preprocessing import OneHotEncoder 
    oh=OneHotEncoder()
    dum=pd.get_dummies(df,columns=strleb)
    return dum
def mima(df):
    #transforming the data
    from sklearn.preprocessing import MinMaxScaler
    tr=MinMaxScaler
    tr.fit(df[df.columns[:-1]],df[df.columns[-1]])
    dumtr=tr(df)
    return dumtr
```

```python
#all columns with string entries
strcol=stringcol(df)
#count of entries in strcols with name
cou(df,strcol)
```
